```python
import pandas as pd
import emoji
import string
import re
```

```python
df = pd.read_csv("goud_articles.csv")
df.head()
```

```python
df = df.drop(["url", "published_time", "modified_time"], axis=1)
```

```python
print(f"Total Observations (rows): {df.shape[0]}")
print(f"Total Features (columns): {df.shape[1]}")
```

```python
print(df.dtypes)
```

```python
#Check for Duplicate Values
if (len(df[df.duplicated()]) > 0):
    print("Duplicate values found in the data.")
    df.drop_duplicates(inplace=True)
else:
    print("No duplicate values detected in the data.")
```

```python
#Check null values
missing = df.isnull().sum()
print("Missing Values : ", missing[missing > 0])
```

```python
df.dropna(inplace=True)
```

```python
df["category"].unique()
```

```python
"""
Ponctuation, emojis (dictionnaire en arabe), special characters, code switching, urls (suppression)
normalisation (7 9 3 -> ح ق ع )
tachkil

"""

features = {
    # "SpecialCharacters": set(),
    "Ponctuations": set(),
    "Emojis": set(),
    "URLs": set(),
}

def extract_features(message: str, features: dict):

    # Emojis
    emojis = emoji.emoji_list(message)
    features["Emojis"].update([d["emoji"] for d in emojis])

    # URLs
    url_pattern = r'(https?://\S+|www\.\S+)'
    urls = re.findall(url_pattern, message)
    features["URLs"].update(urls)

    # Ponctuations
    for c in message:
        if c in string.punctuation:
            features["Ponctuations"].add(c)

df["content"].apply(extract_features, args=(features,))

for key, value in features.items():
    print(f"{key}: {value}")
```
